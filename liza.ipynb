```python
import pandas as pd
from  scipy.stats import mannwhitneyu

df = pd.read_excel('cashback.xlsx')

msc = df[df['регион_проживания'] == 'Москва']
df = df[df['регион_проживания'] != 'Москва']

counts = df['регион_проживания'].value_counts().sort_values(ascending=[False])
indexes = counts.index

regs1 = pd.DataFrame(index=[], columns={'count': []})
regs2 = pd.DataFrame(index=[], columns={'count': []})

for i in range(len(counts)):
    if regs1['count'].sum() <= regs2['count'].sum():
        regs1.at[indexes[i], 'count'] = counts[i]
    else:
        regs2.at[indexes[i], 'count'] = counts[i]

df = pd.read_excel('absolute.xlsx')
df = df.fillna(0)

msc = df[df['регион'] == 'Москва'].T.drop('регион').rename({27: 'Москва'}, axis=1)
regs11 = df[df['регион'].isin(regs1.index)].drop(False, axis=1).groupby('регион').first().T
regs22 = df[df['регион'].isin(regs2.index)].drop(False, axis=1).groupby('регион').first().T

for i in range(34):
    row =regs11.iloc[i]
    mean = row[row.values != 0].median()
    regs11.iloc[i][regs11.iloc[i].values == 0] = mean
    row =regs22.iloc[i]
    mean = row[row.values != 0].median()
    regs22.iloc[i][regs22.iloc[i].values == 0] = mean

regs1 = regs1.median(axis=1).to_frame().rename({0: 'Регионы 1'}, axis=1)
regs2 = regs2.median(axis=1).to_frame().rename({0: 'Регионы 2'}, axis=1)

df = pd.concat([msc, regs1, regs2], axis=1)
df = df.T
df
```
